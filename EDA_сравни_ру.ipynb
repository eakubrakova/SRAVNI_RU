# Исследование маркетинговых каналов

## Анализ распределения клиентов по маркетинговым каналам

```python
channel_counts = adv_df['source_medium'].value_counts()

# Вывод результатов
print("Количество клиентов, привлеченных через каждый маркетинговый канал:")
print(channel_counts)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=channel_counts.index, y=channel_counts.values)
plt.title('Количество клиентов, привлеченных через каждый маркетинговый канал')
plt.xlabel('Маркетинговый канал')
plt.ylabel('Количество клиентов')
plt.xticks(rotation=45)
plt.show()
```
